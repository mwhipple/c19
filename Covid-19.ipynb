### Cases

Here is information about local confirmed cases. This is being treated as a very coarse approximation of risk. Some reasons to likely not read too much in to this data are:

* The incubation period
* Unreported cases
* False positives/negatives
* Testing availability/volume
* The graph of transmission is not likely to clearly correspond to geographic locality

Many of these concerns are likely to be explored and substantiated (or not) later.

Likely the single biggest dependent input variable is testing volume/availability: operating under the assumption that the level of testing does not decrease then the number of cases can serve as a benchmark which can at least be considered relative to previous instances of itself: which is the initial assumption under which I will be operating (and attempting to verify).


```python
import pandas as pd

df = pd.read_csv("cases.csv")
df['change'] = df['confirmed'].diff()
df
```
